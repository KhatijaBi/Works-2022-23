```python
#con.execute("DROP TABLE Assets")
sql = """
CREATE TABLE IF NOT EXISTS Assets (
  Access_No INTEGER PRIMARY KEY,
  Asset_ID Int,         -- for now, refers to book
  Dept_ID Int,
  Call_No Text,         -- location on the shelf 
  Asset_Code Text,      -- inventory number, not used now
  is_physical INTEGER DEFAULT 1 CHECK(is_physical in (0, 1)) , -- physical or digital
  can_issue INTEGER DEFAULT 1 CHECK(can_issue in (0, 1)) ,   
  available INTEGER DEFAULT 1 CHECK(available in (0, 1)),      -- updated with issue/return
  Foreign Key(Dept_ID) References Department(Dept_ID),
  Foreign Key(Asset_ID) References Books(Book_ID)  
);"""

con.execute(sql)
pd.read_sql("PRAGMA table_info(Assets)", con)
```

```python
def book_records():
    con = lite.connect('db_Library_ref.sqlite')

    df = pd.read_excel('Library_data.xlsx')
    col = ['Category', 'Access_No', 'Title', 'Authors', 'Asset_Code', 'Department',
           'Subject', 'Edition', 'Date_Year', 'Call_No', 'Publisher', 'Place',
           'ISBN', 'Price', 'Bill_Date', 'Vendor', 'Pages']

    book_df = df[['Access_No','Title','ISBN','Edition','Date_Year','Pages','Subject','Publisher','Authors']]
    book_df = book_df.fillna('')

    book_list = []
    assn_list = []
    gb_isbn = book_df.groupby('ISBN')
    for name, group in gb_isbn:     
        row = group.iloc[0]

        # subject key
        key = row['Subject']
        get_subject_id = f"""
          SELECT Subject_ID FROM Subject 
          WHERE Subject.Subj_Name='{key}';
        """
        Subj_ID = con.execute(get_subject_id).fetchone()[0]

        # publisher key
        key = row['Publisher']
        get_publ_id = f"""
          SELECT Publ_ID FROM Publisher 
          WHERE Publisher.Publ_Name='{key}';
        """
        try: 
            Publ_ID = con.execute(get_publ_id).fetchone()[0]
        except:
            Publ_ID = 117  # Other

        try:
            year = int(row['Date_Year'])
        except:
            pass

        isb = str(row['ISBN'])
        title = row['Title']
        edition = row['Edition']
        pages = row['Pages']
        book_list.append([isb, title, Subj_ID, edition, Publ_ID, year, pages])

        # author key     
        key = row['Authors']
        get_auth_id = f"""
          SELECT Author_ID FROM Author 
          WHERE Author.Author like'{key}%';
        """
        # author_ID
        try:
            auth_ID = con.execute(get_auth_id).fetchone()[0]
        except:
            auth_ID = 2259  # Others

        bk_id = row['ISBN']

        assn_list.append([auth_ID, bk_id])

    con.close()    
    return book_list, assn_list
```

```python
def asset_data():
    df = pd.read_excel('Library_data.xlsx')
    asset_data = []
    with lite.connect('db_Library_ref.sqlite') as con:
        for index, row in df.iterrows():
            if row['ISBN']:
                # get Dept_ID
                key = row['Department']
                sql_dept = f"""
                  SELECT Dept_ID FROM Department  
                  WHERE Dept = '{key}'
                """
                dept_id = con.execute(sql_dept).fetchone()[0]            

                asset_data.append([row['Access_No'], row['ISBN'], dept_id, row['Call_No'], row['Asset_Code']])

    return asset_data
```
